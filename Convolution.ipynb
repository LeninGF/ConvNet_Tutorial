```python
from google_images_download import google_images_download   #importing the library

response = google_images_download.googleimagesdownload()   #class instantiation

arguments = {"keywords":"guitar","limit":1,"print_urls":True, 
             "format":"jpg", "output_directory": "tmp_img" }   #creating list of arguments
paths = response.download(arguments)   #passing the arguments to the function
print(paths)   #printing absolute paths of the downloaded images
```

```python
paths
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
from   tensorflow.keras.preprocessing.image import img_to_array, load_img
img_path = paths.get('guitar')
img_path = img_path[0]

img = load_img(img_path, target_size=(400, 400))
# print(img_to_array(img))
plt.figure()
plt.imshow(img_to_array(img)/255)
plt.show()
```

```python
x   = img_to_array(img)                           # Numpy array with shape (150, 150, 3)
x   = x.reshape((1,) + x.shape)                   # Numpy array with shape (1, 150, 150, 3)

# Rescale by 1/255
x /= 255.0
print(x.shape)
```

```python
import tensorflow as tf
# x = tf.reshape(x, shape=[-1,400,400,3])
y = tf.nn.conv2d(x, filter=[3,3,1,1], strides=[1,1,1,1], padding='SAME')
```
